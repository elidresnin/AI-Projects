```python
import numpy as np

stocks = np.array(df).tolist()
print(stocks)

columns = df.columns.values.tolist()
print(columns)
print(columns.index('sector'))
```

```python
# Create a list that contains the
# 0 - sector,
# 1 - long business summary,
# 2 - city,
# 3 - state,
# 4 - country,
# 5 - website,
# 6 - industry,
# 7 - profit margins,
# 8 - gross margins,
# 9 - revenue growth,
# 10 - target low price,
# 11 - recommendation key,
# 12 - target median price,
# 13 - current price,
# 14 - number of analyst opinions,
# 15 - target high price,
# 16 - total cash,
# 17 - total debt,
# 18 - total cash per share,
# 19 - fiftytwoweekhigh,
# 20 - fiftytwoweeklow, and
# 21 - regular market price.
stonks = [[f[1], f[3], f[4],f[6], f[7], f[9], f[12], f[14], f[15], f[17], f[20], f[21], f[24], f[25], f[29], f[33], f[34], f[35], f[37], f[141],f[144], f[150]] for f in stocks if not m.isnan(f[24])]
```

### **Map**

```python
# truncate each long business summary so that it is only one setence long. (Keep only the first sentence)
def truncate(stonk):
    new_str = ""
    for letter in stonk[1]:
        if letter != '.':
            new_str += letter
        else:
            break

    stonk[1] = new_str + "."
    return stonk


stonks = list(map(truncate, stonks))



print(stonks)
```

```python
# round each current price to the nearest dollar.
rounded = list(map(round, [stock[13] for stock in stonks]))

for x in range(len(stonks)):
    stonks[x][13] = rounded[x]

print(stonks)
```

```python
# lower case each sector.
lower = list(map(str.lower, [stock[0] for stock in stonks]))

for x in range(len(stonks)):
    stonks[x][0] = lower[x]

print(stonks)
```

### **Filter**

```python
# Remove any row that has less than 10 analyst opinions.
print(len(stonks))
stonks = list(filter(lambda x: x[14] >= 10, stonks))
print(len(stonks))

print(stonks)
```

```python
# Remove any row that does not have a recommendation key of buy.

print(len(stonks))
stonks = list(filter(lambda x: x[11] == 'buy', stonks))
print(len(stonks))

print(stonks)
```

```python
# Remove any row that has greater total debt than total cash.

print(len(stonks))
stonks = list(filter(lambda x: x[17] > x[16], stonks))
print(len(stonks))

print(stonks)
```

```python
# Remove any row that has a below average revenue growth. (Find the average for the stocks that are left in the list after prevous removals.)

revGrowth = [x[9] for x in stonks]

print(revGrowth)
avg = np.average(revGrowth)
print(avg)

print(len(stonks))
stonks = list(filter(lambda x: x[9] < avg, stonks))
print(len(stonks))

print(stonks)
revGrowth = [x[9] for x in stonks]
print(np.average(revGrowth))
```

```python
# Find all stocks from the current list that are in the technology sector.
tech_stocks = list(filter(lambda x: x[0] == "technology", stonks))

print(tech_stocks)
```

```python
# Find all stocks from the current list that are in the healthcare sector.
print(list(filter(lambda x: x[0] == "healthcare", stonks)))
```

### **Reduce**

```python
# Use the reduce function to find the sum of the current price for all technology sector stocks then use this sum to find the average.
from functools import reduce
print((reduce(lambda x, y: x + y, [x[13] for x in tech_stocks]))/len(tech_stocks))
```
